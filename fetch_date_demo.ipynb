```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas
import requests
import sys
import numpy as np
from pywtk.site_lookup import get_3tiersites_from_wkt
WTK_URL = "https://h2oq9ul559.execute-api.us-west-2.amazonaws.com/dev"
from pywtk.wtk_api import get_nc_data_from_url
```

```python
# 确定目标位置的经纬度矩形区域
min_lng = -78
max_lng = -66
min_lat = 38
max_lat = 39
wkt = "POLYGON(({0} {3},{1} {3},{1} {2},{0} {2},{0} {3}))".format(min_lng, max_lng, min_lat, max_lat)
```

```python
# 获得风电场详细数据
site_df = get_3tiersites_from_wkt(wkt)
```

```python
# 取52个风电机组的id
site_id_list = list(site_df.index[:52])
```

```python
# 获取52个机组出力数据
met_data = pandas.DataFrame()
for site_id in site_id_list:
    start = pandas.Timestamp('2008-01-01', tz='utc')
    end = pandas.Timestamp('2014-01-01', tz='utc')
    utc = True
    attributes = ["power"]
    met_data_temp = get_nc_data_from_url(WTK_URL+"/met", site_id, start, end, attributes, utc=utc)
    met_data = met_data.join(met_data_temp,how='outer',rsuffix='_'+str(site_id))
met_data.columns = [ 'power_' + str(i) for i in site_id_list]
```
